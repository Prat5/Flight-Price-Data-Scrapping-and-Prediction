```python
def data_ready_to_scrap(departure_destination,arrival_destination):
    driver.get('https://www.google.com/travel/flights?tcfs&ved=2ahUKEwjxzpjlmPn4AhW4wYQAHcloA_sQyJABegQIABAX&ictx=2')
    time.sleep(2)
    driver.find_element_by_xpath('/html/body/c-wiz[2]/div/div[2]/c-wiz/div/c-wiz/div[2]/div[1]/div[1]/div[1]/div/div[1]/div[1]/div/div[1]/div[1]/div/button/span[2]').click()
    driver.find_element_by_xpath('/html/body/c-wiz[2]/div/div[2]/c-wiz/div/c-wiz/div[2]/div[1]/div[1]/div[1]/div/div[1]/div[1]/div/div[1]/div[2]/div[2]/ul/li[2]').click()
    time.sleep(1)
    driver.find_element_by_xpath('/html/body/c-wiz[2]/div/div[2]/c-wiz/div/c-wiz/div[2]/div[1]/div[1]/div[1]/div/div[2]/div[1]/div[1]/div/div/div[1]/div/div/input').clear()
    driver.find_element_by_xpath('/html/body/c-wiz[2]/div/div[2]/c-wiz/div/c-wiz/div[2]/div[1]/div[1]/div[1]/div/div[2]/div[1]/div[1]/div/div/div[1]/div/div/input').click()
    driver.find_element_by_xpath('/html/body/c-wiz[2]/div/div[2]/c-wiz/div/c-wiz/div[2]/div[1]/div[1]/div[1]/div/div[2]/div[1]/div[6]/div[2]/div[2]/div[1]/div/input').send_keys(departure_destination)
    time.sleep(1)
    driver.find_element_by_xpath('/html/body/c-wiz[2]/div/div[2]/c-wiz/div/c-wiz/div[2]/div[1]/div[1]/div[1]/div/div[2]/div[1]/div[6]/div[3]/ul/li[1]/div[2]').click()
    time.sleep(0.5)
    driver.find_element_by_xpath('/html/body/c-wiz[2]/div/div[2]/c-wiz/div/c-wiz/div[2]/div[1]/div[1]/div[1]/div/div[2]/div[1]/div[4]/div/div/div[1]/div/div/input').click()
    driver.find_element_by_xpath('/html/body/c-wiz[2]/div/div[2]/c-wiz/div/c-wiz/div[2]/div[1]/div[1]/div[1]/div/div[2]/div[1]/div[6]/div[2]/div[2]/div[1]/div/input').send_keys(arrival_destination)
    time.sleep(1)
    driver.find_element_by_xpath('/html/body/c-wiz[2]/div/div[2]/c-wiz/div/c-wiz/div[2]/div[1]/div[1]/div[1]/div/div[2]/div[1]/div[6]/div[3]/ul/li[1]/div[2]').click()
    time.sleep(2)
    driver.find_element_by_xpath('//*[@id="yDmH0d"]/c-wiz[2]/div/div[2]/c-wiz/div/c-wiz/div[2]/div[1]/div[1]/div[1]/div/div[2]/div[2]/div/div/div[1]/div/div/div[1]/div/div[1]/div/input').click()
    driver.find_element_by_xpath('/html/body/c-wiz[2]/div/div[2]/c-wiz/div/c-wiz/div[2]/div[1]/div[1]/div[1]/div/div[2]/div[2]/div/div/div[2]/div/div[2]/div[2]/div/div/div[1]/div/div/div[1]/div[3]/div[3]/div[6]/div').click()
    time.sleep(2)
    driver.find_element_by_xpath('/html/body/c-wiz[2]/div/div[2]/c-wiz/div/c-wiz/div[2]/div[1]/div[1]/div[1]/div/div[2]/div[2]/div/div/div[2]/div/div[3]/div[3]/div/button').click()
    time.sleep(3)
    driver.find_element_by_xpath('/html/body/c-wiz[2]/div/div[2]/c-wiz/div/c-wiz/div[2]/div[1]/div/div[4]/div/div/div[2]/div[1]/div/div[1]/span/button/span').click()
    time.sleep(1)
    driver.find_element_by_xpath('/html/body/c-wiz[2]/div/div[2]/c-wiz/div/c-wiz/div[2]/div[1]/div/div[4]/div/div[2]/div[3]/div/div[1]/section/div[2]/div[1]/div/div/div[3]/div/input').click()
    time.sleep(2)
    driver.find_element_by_xpath('/html/body/c-wiz[2]/div/div[2]/c-wiz/div/c-wiz/div[2]/div[1]/div/div[4]/div/div[2]/div[3]/div/div[1]/section/div[1]/span/button').click()
    # Load more flights
    time.sleep(1)
    for i in range(0,10):
        driver.execute_script("window.scrollTo(0, window.scrollY + 2000)")
        time.sleep(0.2)
    driver.find_element_by_xpath('//button[@class="VfPpkd-LgbsSe VfPpkd-LgbsSe-OWXEXe-k8QpJ VfPpkd-LgbsSe-OWXEXe-Bz112c-M1Soyc VfPpkd-LgbsSe-OWXEXe-dgl2Hf nCP5yc AjY5Oe LQeN7 nJawce OTelKf iIo4pd"]').click()
    time.sleep(5)
```

```python
def feature_extration():
    for i in range(0,61):
        main = driver.find_elements_by_xpath('//div[@class="TQqf0e sSHqwe tPgKwe ogfYpf"]')
        for i in main:
            airline =i.text
            Airline.append(airline)
            date = driver.find_element_by_xpath('/html/body/c-wiz[2]/div/div[2]/c-wiz/div/c-wiz/div[2]/div[2]/div[2]/div[1]/div/div[1]/label[1]/span[1]/div').text
            Date.append(date)

        main = driver.find_elements_by_xpath('//div[@class="zxVSec YMlIz tPgKwe ogfYpf"]')
        for i in main:
            try:
                departure = i.text.split('\n')[0]
                Departure.append(departure)
            except:
                Departure.append('-')
            try:
                arrival = i.text.split('\n')[2]
                Arrival.append(arrival)
            except:
                Arrival.append('-')
                

        main = driver.find_elements_by_xpath('//div[@class="gvkrdb AdWm1c tPgKwe ogfYpf"]')
        for i in main:
            try:
                duration = i.text
                Duration.append(duration)
            except:
                Duration.append('-')

        main = driver.find_elements_by_xpath('//span[@class="z0fuv sSHqwe tPgKwe ogfYpf"]')
        for i in main:
            try:
                route = i.text
                Route.append(route)
            except:
                Route.append('-')

        main = driver.find_elements_by_xpath('//div[@class="BbR8Ec"]')
        for i in main:
            try:
                stop = i.text.split('\n')[0]
                Stops.append(stop)
            except:
                Stops.append('-')
            try:
                layover = i.text.split('\n')[1].split(' ')[-1]
                Layover.append(layover)
            except:
                Layover.append("-")
            try:
                layover_time = i.text.split('\n')[1].split(' ')[:-1]
                layover_time = ' '.join(layover_time)
                Layover_Time.append(layover_time)
            except:
                Layover_Time.append('-')

        main = driver.find_elements_by_xpath('//div[@class="y0NSEe V1iAHe tPgKwe ogfYpf"]')
        for i in main:
            try:
                co2 = i.text.split('\n')[0]
                Co2.append(co2)
            except:
                Co2.append('-')
            try:
                emission = i.text.split('\n')[1]
                Emissions.append(emission)
            except:
                Emissions.append("-")

        main = driver.find_elements_by_xpath('//div[@class="U3gSDe"]')
        for i in main:
            try:
                price = i.text
                Price.append(price)
            except:
                Price.append('-')

        ref = driver.current_url
        driver.get(ref)
        try:
            driver.find_element_by_xpath('/html/body/c-wiz[2]/div/div[2]/c-wiz/div/c-wiz/div[2]/div[1]/div/div[2]/div[2]/div/div/div[1]/div/div/div[1]/div/div[1]/div/div[3]/button').click()
        except:
            ref = driver.current_url
            driver.get(ref)
            pass
        time.sleep(6)
```
